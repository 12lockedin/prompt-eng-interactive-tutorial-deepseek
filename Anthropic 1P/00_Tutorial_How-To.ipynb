### The OpenAI SDK & DeepSeek Compatibility
We will be using the [OpenAI python SDK](https://platform.openai.com/docs/api-reference) since DeepSeek's API is designed to be compatible with OpenAI's format, making it easy to integrate if you're already familiar with OpenAI's SDK.

Below is an example of what running a prompt will look like in this tutorial. First, we create `get_completion`, which is a helper function that sends a prompt to DeepSeek and returns the generated response. Run that cell now.

```python
from openai import OpenAI

client = OpenAI(
    api_key=API_KEY,
    base_url=BASE_URL
)

def get_completion(prompt: str):
    response = client.chat.completions.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
```

Now we will write out an example prompt for DeepSeek and print the output by running our `get_completion` helper function. Running the cell below will print out a response beneath it.

Feel free to play around with the prompt string to elicit different responses.

```python
# Prompt
prompt = "Hello, DeepSeek!"

# Get response
print(get_completion(prompt))
```

### Using the Reasoning Model

DeepSeek also offers a specialized reasoning model called `deepseek-reasoner` which provides Chain-of-Thought (CoT) capabilities. If you want to experiment with this model, here's how you can access both the reasoning process and the final answer:

```python
def get_reasoned_completion(prompt: str):
    response = client.chat.completions.create(
        model="deepseek-reasoner",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.0,
        max_tokens=2000
    )
    
    reasoning_content = response.choices[0].message.reasoning_content
    final_answer = response.choices[0].message.content
    
    print("Reasoning process:")
    print(reasoning_content)
    print("\nFinal answer:")
    print(final_answer)
    
    return final_answer

get_reasoned_completion("Hey Deepseek! Here's a challenge for you: DPU IM JPZNTR. Cue: Qwerty cipher and in spanish!")
```
